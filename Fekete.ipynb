Given a pair of numbers $(n,p)$ where $p$ is a prime, the following function tells whether $n$ is a quadratic residue modulo $p$.

```python
def quadratic_residue(n,p):
    #given a number and a prime p, compute (n/p)
    if n %p==0:
        return 0
    else: 
        for j in range(1,p-1):
            if (n-j**2)%p==0:
                return 1
    return -1  
```

The following program returns the set of all quadratic/non-quadratic residue modulo $p$.

```python
def quadratic_list(p):
    #given a prime number p, list of quadratic/non-quadratic residue modulo p
    my_list=[]
    for i in range(1,p):
        my_list.append(quadratic_residue(i,p))
    return my_list    
```

The following function computes all quadratic residue modulo $p$.

```python
def square(p):
    #given a prime number p, list all quadratic residue modulo p 
    my_list=[]
    for i in range(1,p):
        if quadratic_residue(i,p)==1:
            my_list.append(i)
    return my_list        
```

# The case $p \equiv 3 \pmod{4}$

Next we write the function that compute the coeffcients of 
$$f_p(x)=\sum_{k=0}^{p-3} a_{k}x^{k}$$

Using the relation 
$$ \sum_{r=1}^{p-1} \left(\frac{r}{p} \right)x^{r-1}=(1-x)f_p(x). $$

We can see that for $3 \leq k \leq p$ we have 

$$a_{p-k}=\sum_{i=1}^{k-2} \left(\frac{i}{p} \right). $$


```python
p=19
name= quadratic_list(p)
print(' '.join(str(x) for x in name))
```

```python
The following functions computes the coefficients of Fekete polynomials. 
```

```python
def coefs_fekete(p):
    #given a prime p, compute the coefficients of Fekete polynomial 
    quadratic=quadratic_list(p)
    coefs=[]
    for k in range(0,p-2):
        new_term=sum(quadratic[:k+1])
        coefs.append(new_term)
    return coefs    
    
```

# Roots of $F_p(x)$ modulo $p$. 


```python
def values(p,a):
    #compute the values of f_p(a) modulo p. 
    s=0
    for r in range(1,p):
        s+= (quadratic_residue(r,p)*a**(r-1))%p
    return s%p    
```

```python
def roots(p):
    #compute the root sof f_p modulo p 
    result=[]
    for a in range(1,p):
        if values(p,a)==0:
            if a<p/2: 
                result.append(a)
            else:
                result.append(a-p)    
    return result        
```

```python
def value_derivative(p,a):
    #compute the derivative of f_p 
    derivative=0
    for r in range(2,p):
        derivative+= (quadratic_residue(r,p)*(r-1)* a**(r-2))%p
    d=derivative%p          
    v=values(p,a)%p
    result=d**2+v**2             
    return [v,d] 

def double_roots(p):
    #test whether f_p has double roots 
    result=[]
    for a in range(1,p):
        if (value_derivative(p,a)[0]==0) and (value_derivative(p,a)[1]==0) :
            if a<p/2: 
                result.append(a)
            else:
                result.append(a-p)    
    return result        
```
